# Problem Çözme ve Etkili Sorgulama
# Modul 1

**Yapılandırılmış Düşünme:**

- Yapılandırılmış düşünme, bir problemi daha küçük adımlara ayırma ve bu adımları mantıklı bir çözüme yönlendirme becerisi kazandırır.
- Adımlar: Mevcut durumu tanıma, bilgileri düzenleme, boşlukları ve fırsatları belirleme, seçenekler üretme.

## Giriş

Bu vaka çalışması, **Anywhere Gaming Repair** adında küçük bir işletmenin karşılaştığı benzersiz bir sorunu çözmek için veri analitik sürecinin nasıl kullanıldığını anlatıyor. İşletme, bozuk video oyunu sistemlerini ve aksesuarlarını onaran bir hizmet sunuyor. Sahibi, işini büyütmek için reklam stratejileri geliştirmek istiyordu, ancak hangi yöntemin hedef kitlesine en uygun olduğunu bilmiyordu.

## Reklam Seçenekleri ve Hedef Kitle

Reklam seçenekleri arasında:
- Basılı reklamlar
- Reklam panoları
- TV reklamları
- Toplu taşıma posterleri
- Podcast ve radyo reklamları

**Hedef kitleyi anlamak**, doğru reklam stratejisi seçmenin kritik bir adımıdır. Örneğin:
- Tıbbi ekipman üreticileri, sağlık dergilerini tercih edebilir.
- Yemek şirketleri, aşçılık okullarına yakın poster reklamları düşünebilir.

Bu bağlamda, reklam stratejisi oluştururken iki önemli faktör öne çıkıyor:
1. **Hedef Kitle**: Reklamın kimlere ulaşması gerektiği.
2. **Bütçe**: Reklam türünün maliyeti.

## Veri Analizi Süreci

### 1. **Sorunun Tanımlanması**
Maria, problemin ne olduğunu belirlemek için paydaşlarla çalıştı. Sorun:
- Hedef kitlenin net olarak tanımlanmaması.
- Hangi reklam türünün en etkili olacağının bilinmemesiydi.

### 2. **Veri Toplama**
Maria, şu verileri topladı:
- Video oyun sistemleri sahipliği ile ilgili bilgiler.
- Farklı reklam yöntemlerinin popülerliği ve etkisi.

### 3. **Veri Temizleme**
Verilerdeki hataları ve yanlışlıkları ortadan kaldırdı:
- Eksik bilgiler tamamlandı.
- Aykırı değerler çıkarıldı.

### 4. **Veri Analizi**
Maria şu sorulara cevap aradı:
1. Video oyun sistemlerine sahip olma olasılığı en yüksek olan kişiler kimlerdir?
2. Bu insanlar nerede ve hangi reklam yöntemlerini görme olasılıkları daha yüksektir?

**Sonuçlar:**
- 18-34 yaş arası bireyler, video oyun sistemleri ile ilgili satın alma yapma olasılığı en yüksek gruptur.
- Podcast ve TV reklamları, bu kitlenin dikkatini çekmek için uygundur.

### 5. **Sonuçların Paylaşılması**
Maria, analiz sonuçlarını net ve çekici görsellerle sundu. Önerisi:
- TV reklamlarının maliyetinin yüksek olması nedeniyle, daha uygun maliyetli olan **podcast reklamlarını** tercih etmek.

### 6. **Aksiyon**
Anywhere Gaming Repair, yerel bir podcast ajansıyla çalışarak 30 saniyelik bir reklam hazırladı. Bu reklam, haftada birkaç kez yayınlandı. Sonuçlar:
- İlk hafta müşteri artışı gözlemlendi.
- 4 hafta sonunda 85 yeni müşteri kazanıldı.

## Sonuç

Bu vaka çalışması, veri analizi süreçlerinin nasıl etkili problem çözme yöntemleri sunduğunu göstermektedir. **Altı aşamalı veri analizi süreci**:
1. Sorunu tanımla.
2. Veriyi topla.
3. Veriyi temizle.
4. Veriyi analiz et.
5. Sonuçları paylaş.
6. Aksiyon al.

Bu yaklaşım, gerçek dünya sorunlarının çözümünde nasıl kullanılabileceğine dair güçlü bir örnek sunmaktadır.

**Adım 1: Sorun(Ask)**
Ne olduğunu bilmediğiniz bir sorunu çözmeniz mümkün değildir. Bunlar dikkate alınması gereken bazı hususlardır:

Çözmeye çalıştığınız sorunu tanımlayın 

Paydaşın beklentilerini tam olarak anladığınızdan emin olun

Asıl soruna odaklanın ve dikkatinizi dağıtacak şeylerden kaçının

Paydaşlarla işbirliği yapmak ve açık bir iletişim hattı oluşturmak

Bir adım geri atın ve tüm durumu bağlam içinde görün

Bu adımda kendinize sormanız gereken sorular: 
Paydaşlarım sorunlarının ne olduğunu söylüyor?

Sorunları belirlediğime göre, paydaşların sorularını çözmelerine nasıl yardımcı olabilirim?


**Adım 2: Hazırlık(Prepare)**
Sorularınızı yanıtlamak için hangi verileri toplamanız gerektiğine ve bu verileri yararlı olacak şekilde nasıl düzenleyeceğinize karar vereceksiniz. Karar vermek için iş görevinizi kullanabilirsiniz:

Hangi metriklerin ölçüleceği

Veritabanınızdaki verileri bulun

Bu verileri korumak için güvenlik önlemleri oluşturun

Bu adımda kendinize sormanız gereken sorular:
Bu sorunu nasıl çözeceğimi bulmak için neye ihtiyacım var?

Hangi araştırmayı yapmam gerekiyor?



**3. Adım: Süreç(Process)**
Temiz veri en iyi veridir ve olası hatalardan, yanlışlıklardan veya tutarsızlıklardan kurtulmak için verilerinizi temizlemeniz gerekecektir. Bu şu anlama gelebilir:

Yanlış girilmiş verileri bulmak için elektronik tablo işlevlerini kullanmak 

Fazladan boşluk olup olmadığını kontrol etmek için SQL işlevlerini kullanma

Tekrarlanan girişleri kaldırma

Verilerdeki yanlılığı mümkün olduğunca kontrol etmek

Bu adımda kendinize sormanız gereken sorular:
Hangi veri hataları veya yanlışlıkları çözmeye çalıştığım soruna mümkün olan en iyi yanıtı almamı engelleyebilir?

Elimdeki bilgilerin daha tutarlı olması için verilerimi nasıl temizleyebilirim?



**Adım 4:  Analiz(Analyze)**
Verileriniz hakkında analitik düşünmek isteyeceksiniz. Bu aşamada, verilerinizi daha kolay hale getirmek için sıralayabilir ve biçimlendirebilirsiniz:

Hesaplamalar yapmak

Birden fazla kaynaktan gelen verileri birleştirin

Sonuçlarınızla tablolar oluşturun

Bu adımda kendinize sormanız gereken sorular:
Verilerim bana hangi hikayeyi anlatıyor?

Verilerim bu sorunu çözmeme nasıl yardımcı olacak?

Şirketimin ürün veya hizmetine kimin ihtiyacı var? Ne tür bir kişinin bunu kullanma olasılığı daha yüksek?



**Adım 5: Paylaşın(Share)**
Herkes sonuçlarını farklı şekilde paylaşır, bu nedenle sonuçlarınızı grafikler veya gösterge tabloları gibi araçlar aracılığıyla verileri kullanarak analizinizin net ve cazip görselleriyle özetlediğinizden emin olun. Bu, paydaşlarınıza sorunlarını çözdüğünüzü ve bu sonuca nasıl ulaştığınızı gösterme şansınızdır. Paylaşım kesinlikle ekibinize yardımcı olacaktır: 

Daha iyi kararlar alın

Daha bilinçli kararlar alın

Daha güçlü sonuçlara yol açar

Bulgularınızı başarılı bir şekilde iletin

Bu adımda kendinize sormanız gereken sorular:
Paydaşlara sunduğum şeyi nasıl ilgi çekici ve kolay anlaşılır hale getirebilirim?

Dinleyici olsaydım bunu anlamama ne yardımcı olurdu?



**Adım 6: Harekete geçin(Act)**
Şimdi verilerinize göre hareket etme zamanı. Veri analizinizden öğrendiğiniz her şeyi alıp kullanacaksınız. Bu, paydaşlarınıza bulgularınıza dayalı tavsiyeler sunarak veri odaklı kararlar almalarını sağlamak anlamına gelebilir.

Bu adımda kendinize sormanız gereken sorular:
Paylaşım aşamasında (5. adım) aldığım geri bildirimi paydaşın ihtiyaç ve beklentilerini gerçekten karşılamak için nasıl kullanabilirim?

Bu altı adım, veri analizi sürecini daha küçük, yönetilebilir parçalara ayırmanıza yardımcı olabilir, buna yapılandırılmış düşünme denir. Bu süreç dört temel faaliyeti içerir:

Mevcut sorun ya da durumun tanınması

Mevcut bilgilerin düzenlenmesi

Boşlukların ve fırsatların ortaya çıkarılması

Seçeneklerinizi belirleme

Bir veri analisti olarak kariyerinize başlarken, rolünüz ve beklentilerinizle ilgili birkaç farklı yöne çekildiğinizi hissetmeniz normaldir. Burada özetlenenler gibi süreçleri takip etmek ve yapılandırılmış düşünme becerilerini kullanmak sizi tekrar rayına oturtmaya, boşlukları doldurmaya ve tam olarak neye ihtiyacınız olduğunu bilmenize yardımcı olabilir.

### Veri Analistlerinin Karşılaştığı Altı Temel Problem Türü

Veri analizi süreci, şirketlerin sorunlarını çözmek için güçlü problem çözme becerilerini kullanmalarını sağlar. Bu içerikte, veri analistlerinin sıkça karşılaştığı altı temel problem türünü inceleyeceğiz:

1. **Tahminler Yapmak**
Tahmin, gelecekteki olayları öngörmek için mevcut verilerin kullanılmasıdır.
- **Örnek:**  
  Bir hastane, kronik hastalar için sağlık olaylarını tahmin etmek amacıyla uzaktan hasta izleme kullanabilir. Hastaların günlük sağlık verileri, yaşları ve risk faktörleri bir algoritma tarafından analiz edilerek olası sağlık sorunları önceden tahmin edilebilir. Bu da gelecekteki hastane yatışlarını azaltmaya yardımcı olur.
- **Örnek2:**
  Yeni müşteriler kazanmak için en iyi reklam yöntemini bilmek isteyen bir şirket, analistlerin tahminlerde bulunmasını gerektiren bir soruna örnektir. Konum, medya türü ve geçmiş reklamlar sonucunda elde edilen yeni müşteri sayısı hakkında veri sahibi olan analistler gelecekteki sonuçları garanti edemez, ancak hedef kitleye ulaşmak için en iyi reklam yerleşimini tahmin etmeye yardımcı olabilirler.

2. **Kategorize Etmek**
Bilgilerin ortak özelliklere göre gruplara veya kümelere atanmasıdır.
- **Örnek:**  
  Bir üretici, çalışan performansını analiz ederek mühendislik, bakım ve montaj gibi kategorilerde en etkili ve en az etkili çalışanları gruplandırabilir. Bu gruplandırma, farklı görevlerdeki performans farklarını görmeyi kolaylaştırır.
- **Örnek2:**
  Analistlerin bir şeyleri kategorize etmesini gerektiren bir soruna örnek olarak bir şirketin müşteri memnuniyetini artırma hedefi verilebilir. Analistler müşteri hizmetleri çağrılarını belirli anahtar kelimelere veya puanlara göre sınıflandırabilir. Bu, en iyi performans gösteren müşteri hizmetleri temsilcilerinin belirlenmesine veya gerçekleştirilen belirli eylemlerin daha yüksek müşteri memnuniyeti puanlarıyla ilişkilendirilmesine yardımcı olabilir.

3. **Alışılmadık Şeyleri Tespit Etmek**
Normdan farklı verilerin belirlenmesidir.
- **Örnek:**  
  Bir okul sisteminde öğrenci kaydında ani bir artış (örneğin, %30'luk bir sıçrama) tespit edilebilir. Bu artışın sebebi olarak bölgedeki yeni apartman komplekslerinin inşası keşfedilebilir. Bu bilgi, ek kaynak planlamasında kullanılabilir.
- **Örnek2:**
  İnsanların sağlık durumlarını izlemelerine yardımcı olan akıllı saatler satan bir şirket, yazılımlarını olağandışı bir şeyi tespit edecek şekilde tasarlamakla ilgilenebilir. Toplu sağlık verilerini analiz eden analistler, ürün geliştiricilerin belirli veriler normal seyretmediğinde bunları tespit etmek ve alarm vermek için doğru algoritmaları belirlemelerine yardımcı olabilir.

4. **Temaları Belirlemek**
Bilgilerin daha geniş kavramlar altında gruplandırılmasıdır.
- **Örnek:**  
  Çalışan performansı kategorilerden (mühendislik, bakım, montaj) daha geniş kavramlara (düşük üretkenlik, yüksek üretkenlik) gruplanabilir. Bu temalar, düşük üretken çalışanlar için destek planları oluşturmayı ve en iyi performans gösterenleri ödüllendirmeyi mümkün kılar.
- **Örnek2:**
  Kullanıcı deneyimi (UX) tasarımcıları, kullanıcı etkileşimi verilerini analiz etmek için analistlere güvenebilir. Analistlerin bir şeyleri kategorize etmesini gerektiren sorunlara benzer şekilde, kullanılabilirlik iyileştirme projeleri de analistlerin iyileştirme için doğru ürün özelliklerine öncelik verilmesine yardımcı olacak temaları belirlemesini gerektirebilir. Temalar çoğunlukla araştırmacıların verilerin belirli yönlerini keşfetmelerine yardımcı olmak için kullanılır. Bir kullanıcı çalışmasında, kullanıcı inançları, uygulamaları ve ihtiyaçları temalara örnektir. 

Şimdiye kadar bir şeyleri kategorize etmekle temaları belirlemek arasında bir fark olup olmadığını merak ediyor olabilirsiniz. Bunu düşünmenin en iyi yolu şudur: şeyleri kategorize etmek, öğeleri kategorilere atamayı içerir; temaları tanımlamak, bu kategorileri daha geniş temalar halinde gruplandırarak bir adım daha ileri götürür.

5. **Bağlantıları Keşfetmek**
Farklı veriler arasındaki ilişkilerin ve benzerliklerin belirlenmesidir.
- **Örnek:**  
  Bir scooter üreticisi, tekerlek tedarikçisindeki bir sorunu tespit edebilir. Sorunun kauçuk tedarikçisinden kaynaklandığını fark ederek, tüm tarafların iş birliği yapmasıyla çözüm geliştirebilir.
- **Örnek2:**
  Üçüncü taraf bir lojistik şirketinin, sevkiyatların müşterilere zamanında teslim edilmesi için başka bir şirketle birlikte çalışması, analistlerin bağlantıları keşfetmesini gerektiren bir sorundur. Analistler, sevkiyat merkezlerindeki bekleme sürelerini analiz ederek zamanında teslimat sayısını artırmak için uygun program değişikliklerini belirleyebilirler. 

6. **Kalıpları Bulmak**
Geçmiş verilerden yola çıkarak tekrarlayan eğilimlerin belirlenmesidir.
- **Örnek:**  
  E-ticaret şirketleri, yıl boyunca belirli dönemlerde müşteri davranışlarını analiz eder. Örneğin:
  - Kasırga öncesinde konserve gıda satışlarında artış.
  - Yaz aylarında soğuk hava aksesuarlarının daha az satılması.  
  Bu kalıplar, stok yönetimini optimize etmek için kullanılabilir.
- **Örnek2:**
  Makine arızalarından kaynaklanan duruş sürelerinin en aza indirilmesi, analistlerin verilerdeki kalıpları bulmasını gerektiren bir sorun örneğidir. Örneğin, bakım verilerini analiz ederek, düzenli bakımın 15 günden daha fazla gecikmesi durumunda çoğu arızanın meydana geldiğini keşfedebilirler. 

---

## Sonuç

Veri analistleri, her biri farklı bir yaklaşım gerektiren bu altı temel problem türüyle çalışır:
1. Tahminler yapmak.
2. Şeyleri kategorize etmek.
3. Alışılmadık şeyleri tespit etmek.
4. Temaları belirlemek.
5. Bağlantıları keşfetmek.
6. Kalıpları bulmak.

Bu beceriler, bir veri analisti olarak kariyerinizde karşılaşacağınız çeşitli problemleri çözmede önemli bir temel oluşturacaktır. Gerçek dünya örnekleriyle bu kavramların daha iyi anlaşılması sağlanabilir ve bu örnekler gelecekte daha fazla konuşulacaktır.

# Etkili Sorular Sorma: Veri Analizi Sürecinin İlk Adımı

## Giriş

Veri analizi süreci, **doğru sorular sormakla** başlar. Etkili sorular, temel içgörüler elde etmenin ve problem çözmenin anahtarıdır. Bu yazıda, veri analisti olarak doğru soruları nasıl sorabileceğinizi, **SMART metodolojisi**ni nasıl kullanacağınızı ve adil soruların önemini öğreneceksiniz.

## 1. **Etkili Soruların Önemi**
- **Sürekli Sorular Sorma**:  
  Veriler hakkında daha fazla bilgi edinmek için mümkün olduğunca çok soru sormak gerekir. Ne kadar fazla soru sorarsanız, içgörüleriniz o kadar güçlü olur.

- **Etkisiz Soruların Zararları**:
  - **Önde Gelen Sorular**: Cevapları belirli bir şekilde yönlendirir. Örneğin: "Bu şimdiye kadarki en iyi sandviçler, değil mi?"
  - **Kapalı Uçlu Sorular**: Evet veya hayır ile cevaplanır, derin içgörüler sağlamaz.
  - **Belirsiz Sorular**: Yeterince bağlam sağlamaz. Örneğin: "Çikolatayı mı yoksa vanilyayı mı tercih edersiniz?" (Hangi bağlamda?)

## 2. **SMART Metodolojisi**
Etkili soruların oluşturulmasında **SMART metodolojisi** kullanılır. Bu çerçeve, soruları **spesifik**, **ölçülebilir**, **eylem odaklı**, **alakalı** ve **zamana bağlı** hale getirir.

### a. **Spesifik Sorular**(hangi tür veriler kullanıyorsunuz?)
- **Özellikleri**: Basit, anlamlı, tek bir konuya odaklı. Sorular spesifik mi? Sorunu çözmek için yeterince odaklanmış mı? Bağlamı açık mı? 
- **Örnek**:  
  - Kapalı uçlu soru: "Çocuklar bugünlerde yeterince fiziksel aktivite yapıyor mu?"  
  - Spesifik soru: "Çocukların yüzde kaçı haftada en az 5 gün önerilen 60 dakikalık fiziksel aktiviteye ulaşıyor?"

### b. **Ölçülebilir Sorular**
- **Özellikleri**: Cevapları değerlendirilebilir ve sayısal olarak ifade edilebilir. Sorular ölçülebilir mi? Cevaplar size değerlendirilebilir bilgiler sağlıyor mu?
- **Örnek**:  
  - Ölçülemeyen soru: "Yeni bir video neden viral oldu?"  
  - Ölçülebilir soru: "Videomuzun ilk hafta içinde kaç kez paylaşıldığını öğrenebilir miyiz?"

### c. **Eylem Odaklı Sorular**
- **Özellikleri**: Değişim yaratmayı ve harekete geçmeyi sağlar. Sorular bir tür plan tasarlamanıza yardımcı olacak bilgiler sunuyor mu?
- **Örnek**:  
  - Genel soru: "Müşterilerin ürün ambalajımızı geri dönüştürmelerini nasıl sağlayabiliriz?"  
  - Eylem odaklı soru: "Hangi tasarım özellikleri ambalajımızın geri dönüştürülmesini kolaylaştırır?"

### d. **Alakalı Sorular**
- **Özellikleri**: Sorunla doğrudan ilgilidir ve çözüm odaklıdır. Sorular çözmeye çalıştığınız sorunla doğrudan alakalı mı?
- **Örnek**:  
  - Alakasız soru: "Pine Barrens ağaç kurbağalarının kaybolması neden önemli?"  
  - Alakalı soru: "Durham'da 1983-2004 yılları arasında bu kurbağaların kaybolmasına neden olabilecek çevresel faktörler nelerdir?"

### e. **Zamana Bağlı Sorular**
- **Özellikleri**: Zaman aralığını belirtir ve odaklanmayı kolaylaştırır. Cevaplar, belirli bir zaman dilimine odaklanarak analiz edilebilir mi?
- **Örnek**:  
  - Zaman belirtilmeyen soru: "Bu çevresel değişiklikler ne zaman oldu?"  
  - Zamana bağlı soru: "1983-2004 yılları arasında hangi çevresel değişiklikler meydana geldi?"

---

## 3. **Adil Soruların Önemi**
- **Adalet Nedir?**
  - Soruların önyargı yaratmaması veya cevapları yönlendirmemesi.
  - Herkes için anlaşılır ve açık ifadelerle oluşturulması.

- **Önyargılı Soruların Örnekleri**:
  - Önde gelen soru: "Sergilerimizi seviyor musunuz?" (Varsayım içeriyor.)
  - Adil soru: "Sergilerimiz hakkında ne düşünüyorsunuz?"

- **Sonuçlar**:
  - Adil olmayan sorular güvenilmez geri bildirimlere ve değerli içgörülerden yoksun kalmaya yol açar.

# SMART Soruları Hakkında Daha Fazla Bilgi

### Günümüz İş Dünyasında SMART Soruları
Günümüzün birçok sektördeki şirketleri hızlı değişim ve artan belirsizlikle uğraşıyor. Köklü işletmeler bile yeni olanlara ayak uydurma ve sırada ne olacağını bulma baskısı altındadır. Bu durumda doğru soruları sormak, birçok işletmenin yenilikçi fikirleri ateşlemesine yardımcı olabilir. 

Aynı şey veri analitiği için de geçerlidir. Ne kadar bilgiye sahip olursanız olun, doğru sorularla başlamazsanız verileriniz size pek bir şey söylemez. Bu nedenle **SMART soruları**, etkili veri analizi ve problem çözme süreçlerinde kritik bir araçtır.

## SMART Çerçevesinin Gücü
- **Bağlantı Sağlar**: Sorular spesifik olduğu için odaklanmayı kolaylaştırır.

# Modul 2
# Verilerin Karar Vermede Rolü

## Veriler Nedir?
- **Veriler**, bir gerçekler topluluğudur ve karar vermede önemli bir rol oynar.  
- Veri analizi, bu verilerden önemli kalıplar ve içgörüler ortaya çıkarır ve daha bilinçli kararlar almamıza yardımcı olur.

---

## Veri Odaklı ve Veriden İlham Alan Karar Verme
1. **Veri Odaklı Karar Verme**:  
   - Belirli verilere dayanarak karar almayı ifade eder.  
   - Örnek: "Yakınımdaki restoranlar" araması yapıp, restoranları puanlarına göre sıralayarak seçim yapmak.

2. **Veriden İlham Alan Karar Verme**:  
   - Farklı veri kaynaklarını araştırıp ortak noktaları keşfetmeye dayanır.  
   - Örnek: Google, yapay zeka ile yıllarca toplanan verileri analiz ederek veri merkezlerini soğutma maliyetlerini %40'tan fazla azalttı.

---

## Verilerle Gerçek Hayattaki Örnekler
1. **Google İnsan Operasyonları Ekibi**:  
   - Yeni çalışanları işe alma sürecini iyileştirmek için başvuru, görüşme ve oryantasyon verilerini analiz etti.  
   - **Sonuçlar**: 
     - Algoritma kullanarak harika adaylar belirleme.  
     - Yeni çalışanların işe geçişlerini kolaylaştırmak için oryantasyon süreçlerini yeniden yapılandırma.

2. **Verilerle Enerji Tasarrufu**:  
   - Yapay zeka ve toplanan veriler sayesinde veri merkezlerini soğutma enerjisi %40 oranında azaltıldı.

---

## Verilerin Potansiyeli ve Dönüştürülmesi
- **Veri Hacmi**: Dünyadaki verilerin %90'ı sadece son birkaç yılda oluşturulmuştur.  
- **Veriyi Bilgiye Dönüştürmek**:  
   - Veriler yalnızca toplandığında değil, yorumlandığında anlam kazanır.  
   - **Örnek**:  
     - Michael Phelps'in bir yüzme yarışını 1:54 sürede tamamlaması.  
     - Bu süreyi diğer yarışmacılarla karşılaştırarak altın madalya kazandığını anlamak.

---

## Verinin Faydaları
1. **Bağlam ve Analiz**:  
   - Veriler, bağlam olmadan anlamsızdır.  
   - Örnek: Phelps’in süresi ancak diğer verilerle karşılaştırıldığında anlamlı hale gelir.

2. **Bilgi ve Uygulama**:  
   - Veriler analiz edilip bilgiye dönüştürüldüğünde ve uygulandığında en yararlı hale gelir.

---

## Veri Analitiğindeki Sınırlamalar
- Her zaman gerekli verilere erişilemeyebilir.  
- Veriler programlar arasında farklı şekilde ölçülebilir.  
- Bu durum, somut örnekler elde etmeyi zorlaştırabilir.

---

## Sonuç
- Veriler, karar verme süreçlerinde güçlü bir araçtır.  
- Veri analisti olarak işletmelere sorun çözmede ve yeni kararlar almada ihtiyaç duydukları bilgileri sağlayabilirsiniz.  
- Bunun için veri türlerini ve onlarla nasıl çalışacağınızı iyi anlamalısınız.

> "Veriler kendi başına anlamsızdır; ancak analiz ve bağlam ile bilgiye dönüşür."  

> "Unutmayın: Veri karar vermez, ancak kararları iyileştirir."

# Veri Analitiği ve Karar Verme

## Veri Analitiği Uzmanlarının Rolü
Bir veri analitiği uzmanının işi, organizasyonlar için önemli kararları bilgilendirmek adına gerekli verileri sağlamaktır. Uzmanlar, analizlerini iş liderlerinin en iyi kararları almasına yardımcı olacak şekilde çerçevelendirir.

Bu bölümde:
- Verilerin karar alma sürecindeki rolü incelenecek.
- Veri odaklı ve veriden ilham alan kararlar karşılaştırılacak.
- Verilerin projelerin başarısına veya başarısızlığına etkisi örneklerle ele alınacak.

## Veri Odaklı ve Veriden İlham Alan Yaklaşımlar
### Veri Odaklı Kararlar
Veriye dayalı karar verme, iş stratejisini yönlendirmek için gerçeklerin kullanılmasını ifade eder. Bu yaklaşım, mevcut verilerin miktar ve kalitesine bağlıdır. Ancak, yetersiz veya taraflı veriler aşağıdaki riskleri barındırabilir:
- Geçmiş verilere aşırı güven.
- Nitel içgörüleri göz ardı etme.
- Veri toplama ve analizindeki önyargılar.

#### Örnek: A/B Testi
- **Durum:** Bir web sitesi, yeni bir düzenin satışları artıracağını test etmek ister.
- **Yöntem:** Kullanıcıların yarısı eski, yarısı yeni düzeni kullanır.
- **Sonuç:** Yeni düzen daha fazla satış getiriyorsa, yeni düzen kullanılır.

### Veriden İlham Alan Kararlar
Veriden ilham alan kararlar, veri odaklı yaklaşıma daha geniş bir perspektif ekler. Bu yaklaşımda:
- Veriler, duygu ve deneyimlerle desteklenir.
- Ölçülmesi zor niteliksel faktörler dikkate alınır.
- Veri odaklı kararların eğilimli olabileceği tuzaklar önlenebilir.

#### Örnek: Müşteri Memnuniyeti
- **Durum:** Bir destek merkezi, müşteri memnuniyetini artırmayı hedefler.
- **Yöntem:** CSAT puanları analiz edilir, niteliksel geri bildirimler incelenir, çalışanlarla görüşülür.
- **Sonuç:** Hem veriler hem de çalışan girdileriyle strateji belirlenir.

---

## Veri Analizinin Başarı Örnekleri
### PepsiCo
PepsiCo, analitik ve teknolojiyi şu şekilde kullanmıştır:
- Verilerin merkezi bir bulut tabanlı platformda toplanması.
- İşlevler arası iş akışları oluşturulması.
- Verilerin hem dahili hem harici kaynaklarla desteklenmesi.

Bu strateji, müşteri deneyimlerini kişiselleştirmeye ve pazar erişimini genişletmeye olanak sağlamıştır.

---

## Veri Analizinin Başarısızlık Örnekleri
### New Coke Lansman Hatası
- **Durum:** 1985'te New Coke piyasaya sürüldü.
- **Sorun:** Veriler, müşterilerin klasik Coke'un yerini almasıyla ilgili hislerini dikkate almadı.
- **Sonuç:** Şirket, eksik verilerle aldığı karar nedeniyle milyonlarca dolar kaybetti.

### Mars İklim Yörünge Aracı Kaybı
- **Durum:** 1999'da NASA, Mars İklim Yörünge Aracı'nı kaybetti.
- **Sorun:** İngiliz ve Uluslararası Birimler Sistemi arasındaki uyumsuzluk.
- **Sonuç:** Yanlış veri varsayımları nedeniyle araç atmosfere girişte parçalandı.

---

## Anahtar Çıkarımlar
- **Veri Doğruluğu:** Eksiksiz, doğru ve ilgili veriler elde edilmelidir.
- **Karar Süreci:** Veriler doğru bir şekilde yorumlanmalıdır.
- **Analist Rolü:** Analistler, mantık ve yaratıcılığı birleştirerek verilere dayalı stratejiler geliştirmelidir.

### Özet
Bir veri analisti olarak:
- Veri odaklı bir zihniyete sahip olun.
- Çok sayıda soru sorun ve farklı olasılıkları keşfedin.
- Verilerinizi en üst düzeyde özen ve doğrulukla yorumlayın.

# Nicel ve Nitel Veriler: Karar Vermede Verilerin Rolü

## Giriş
Veriler, karar alma süreçlerinde anahtar bir rol oynar. Farklı soru türleri, farklı veri türlerinin oluşturulmasına olanak tanır. Bu videoda, **nicel** ve **nitel** verilerin tanımları, kullanım alanları ve örnekleri ele alınmaktadır.

---

## Nicel Veriler
Nicel veriler, **sayısal gerçeklerin spesifik ve nesnel ölçümleriyle** ilgilidir. Bunlar, genellikle bir problemle ilgili "ne", "kaç" ve "ne sıklıkta" gibi soruları yanıtlar.

### Örnek:
- Her hafta işe giden kaç kişinin trene bindiği gibi ölçülebilir bilgiler.

### Avantajlar:
- Kesinlik ve doğruluk sağlar.
- Çizelgeler veya grafiklerle kolayca görselleştirilebilir.

---

## Nitel Veriler
Nitel veriler, **öznel veya açıklayıcı ölçümleri** ifade eder. Sayısal olarak ölçülemeyen özelliklerin anlaşılmasına yardımcı olur.

### Örnek:
- İnsanların neden belirli bir ünlüyü veya atıştırmalık yiyeceği tercih ettiği.

### Avantajlar:
- **"Neden"** sorularına cevap verir.
- Sorunlara bağlam ekler ve daha derin bir anlayış sağlar.

---

## Nicel ve Nitel Verilerin Birlikte Kullanımı
### İş Örneği: Dondurma Dükkanı
1. **Nicel Veri Analizi:**
   - Kaç tane olumsuz yorum var?
   - Ortalama derecelendirme nedir?
   - Kaç inceleme aynı anahtar kelimeleri içeriyor?(hayal kırıklığı anahtar kelimesi)

2. **Nitel Veri Analizi:**
   - Neden bu kelimeler tekrar ediliyor?
   - Müşterilerin deneyimlerini nasıl geliştirebiliriz?

**Sonuç:**
- İncelemelerde "hayal kırıklığına uğramış" kelimesinin sıkça geçtiği tespit edilir.
- Bu durumun sebebi, popüler lezzetlerin gün içinde tükenmesidir.
- Çözüm olarak, haftalık siparişlerin değiştirilmesi kararı alınır.

### Özet:
- Nicel veri, müşterilerin memnuniyetsizliğini ortaya koyar.
- Nitel veri, memnuniyetsizliğin nedenini açıklar.
- İki veri türü birlikte kullanılarak iş geliştirme stratejisi oluşturulur.

---

## Veri Analisti Olarak Görevleriniz
1. **Doğru Soruları Sormak:**
   - Hangi soruların hangi veri türlerini gerektirdiğini belirleyin.
2. **Paydaşlara Verileri Açıklamak:**
   - Verileri daha iyi anlamalarına yardımcı olacak yaratıcı yöntemler geliştirin.
   - Örneğin: Etkileşimli gösterge tabloları.

---

## Anahtar Çıkarımlar
- **Nicel Veriler:** Ölçülebilir, sayısal gerçeklere dayanır.
- **Nitel Veriler:** Bağlam sağlayan öznel ölçümler içerir.
- Her iki veri türü birlikte kullanıldığında daha etkili sonuçlar elde edilir.
- Sorularınıza bağlı olarak doğru veri türünü seçmek ve analiz etmek, bir veri analisti olarak temel görevinizdir.

### Sonraki Adımlar:
- Etkileşimli gösterge tabloları ve diğer görselleştirme araçlarını öğrenmek.
- Nicel ve nitel verileri birleştiren analizler yaparak iş süreçlerine katkı sağlamak.

# Nicel ve Nitel Veriler: Sinema Salonu Örneği

## Veri Türlerinin Tanımı ve Araçları
### Nicel Veriler
- **Tanım:** Sayısal gerçeklerin spesifik ve nesnel ölçümleridir.
- **Kullanım Alanları:** 
  - Zaman içinde sinemaya katılım.
  - İmtiyaz standının karlılığı.
  - Akşam izleyici tercihleri.
- **Araçlar:**
  - Yapılandırılmış mülakatlar.
  - Anketler.
  - Oylamalar.

### Nitel Veriler
- **Tanım:** Öznel veya açıklayıcı ölçümleri ifade eder.
- **Kullanım Alanları:**
  - İnsanların davranış ve tercih nedenlerini anlamak.
- **Araçlar:**
  - Odak grupları.
  - Sosyal medya metin analizi.
  - Yüz yüze mülakatlar.

---

## Üç Eğilim Üzerine Nicel Veri Analizi
### 1. Zaman İçinde Sinemaya Katılım
- **Geçmiş Veriler:** Son 3 aylık verilere bakıldığında aylık ortalama katılım 550.
- **Tatillerin Etkisi:** Tatillerin olduğu aylarda katılım 1.600'e yükseliyor.
- **Planlama:** Bilet fiyatları arttıktan sonra katılım verilerini tekrar analiz etme kararı alındı.

### 2. İmtiyaz Standının Kârlılığı
- **Mevcut Durum:** Kâr marjı %5'in altında.
- **Satış Verileri:** Ortalama satın alma toplamı 20 $ veya daha az.
- **Çözüm:** Çevrimiçi bir anketle müşteri geri bildirimi toplanarak menü yenileme ve potansiyel kâr artışı sağlanabilir.

### 3. Akşam İzleyici Tercihleri
- **Popüler Saatler:** 7:30 en çok katılımın olduğu saat.
- **Düşük Katılım:** 8:00 gösterimi yerine 8:30 gösterimine geçiş önerisi.
- **Planlama:** Daha geç gösterim saatleri için müşteri önceliklerini anlamak adına anket soruları ekleme.

---

## Nitel Veri Toplama ve Kullanımı
### Anket Soruları
1. **Film Katılımı:**
   - Bugün sinemamızda bir film izleme kararınızda neler etkili oldu?
2. **İmtiyaz Standı Karlılığı:**
   - İmtiyaz standından yaptığınız alışverişlerin kalitesi ve değeri hakkında ne düşünüyorsunuz?
3. **Akşam Gösterim Tercihleri:**
   - Hangi gösterim saatini tercih ediyorsunuz, 8:00 mi yoksa 8:30 mu ve neden?
4. **Bilet Fiyatı Artışı:**
   - Hangi koşullar altında gece gösterimi yerine matineyi tercih edersiniz?

---

## Önemli Çıkarımlar
- **Nicel Veriler:**
  - "Ne oluyor?" sorusuna yanıt verir.
  - Örneğin, katılım veya satış oranlarını ölçer.
- **Nitel Veriler:**
  - "Neden oluyor?" sorusuna yanıt verir.
  - Müşteri tercihlerini ve davranışlarını açıklar.

### Örnek:
- **Nicel Veri:** Müşteriler genellikle saat 7:30’da sinemaya geliyor.
- **Nitel Veri:** Müşteriler 7:30 gösterimini popüler restoranlardan sinemaya rahatça ulaşabildikleri için tercih ediyor.

### Uygulama:
- **Yönetim Kararları:**
  - Daha fazla yatar koltuk eklemek.
  - Daha uygun gösterim saatleri planlamak.
  - Menü yenilikleriyle kârı artırmak.

---

## Sonuç
Nicel ve nitel veriler birlikte kullanıldığında, hem işletme performansını artıran hem de müşteri memnuniyetini iyileştiren kapsamlı stratejiler oluşturulabilir. Bu iki veri türünü doğru bir şekilde analiz etmek, veri analistlerinin değerli içgörüler sağlamasında kritik bir rol oynar.
